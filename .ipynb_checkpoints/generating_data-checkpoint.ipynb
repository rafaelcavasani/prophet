```python
import pandas as pd
import datetime
import random
```

```python
df = pd.DataFrame(columns=['ds', 'y'])
```

```python
time = datetime.datetime(2021, 12, 31, 23, 45)
year_weight = 1
time_range_map = {0: [5,10], 1: [0,5], 2: [0,0], 3: [0,0], 4: [0,0], 5: [0,2], 6: [0,3], 7: [1,4], 8: [5,10], 9: [10, 20], 10: [10,20], 11: [10,20], 12: [20,40], 13: [20,40], 14: [20,30], 15: [20,30], 16: [20,30], 17: [20,25], 18: [30,40], 19: [30,40], 20: [40,50], 21: [30,40], 22: [20,25], 23: [10,15]}
day_weight = {1: 3, 2: 2, 3: 1, 4: 3, 5: 1, 6: 1, 7: 4}
```

```python
# 1 ano -> 35040
# 1/2 ano -> 17376
for i in range(8000):
    time = time + datetime.timedelta(minutes=15)
    time_range = time_range_map[time.hour]
    weight = day_weight[time.today().weekday()]
    value = random.randint(time_range[0], time_range[1]) + weight
    df = pd.concat([pd.DataFrame([[time, value]], columns=df.columns), df], ignore_index=True)
```
